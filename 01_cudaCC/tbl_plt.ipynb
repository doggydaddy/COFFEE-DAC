```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

res=np.array([6.0, 5.5, 5.0, 4.5, 4.0, 3.5, 3.0, 2.5, 2.0])
nrvox=np.array([6267, 8088, 10935, 14856, 21081, 31260, 49877, 86551, 168648])
reqmem=np.array([0.07, 0.12, 0.22, 0.41, 0.83, 1.82, 4.63, 13.95, 52.98])
omp=np.array([7.012, 11.904, 21.1770, 39.180, 79.477, 176.363, 443.851, 1356.103, 4760.729])
cuda=np.array([0.051, 0.085, 0.149, 0.279, 0.584, 1.220, 3.143, 10.892, 121.451])
data=pd.DataFrame({
                   'omp': omp,
                   'cuda': cuda
                   })
data.index=res
print(data)
```

```python

plt.rcParams.update({'font.size': 20})

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10,12), gridspec_kw={'height_ratios': [1, 3]})
fig.suptitle('memory usage and run times')
fig.set_facecolor((0.7, 0.7, 0.7, 0.8))

ax1.plot(res, reqmem, 'o-', linewidth=3)
ax1.set_ylabel('Memory Use (GB)', rotation=90, fontsize=18, labelpad=10)
ax1.invert_xaxis()
ax1.set_facecolor((0.9, 0.9, 0.9))

ax2.plot(res, omp, '.-', color='blue', alpha=0.8, label="OMP", linewidth=1.5)
ax2.plot(res, cuda, '.-', color='green', alpha=0.8, label="CUDA", linewidth=1.5)
ax2.set_xlabel('voxel size (mm)')
ax2.set_ylabel('run time (s)', rotation=90, fontsize=20, labelpad=10)
ax2.invert_xaxis()
ax2.legend(["OMP", "CUDA"])
ax2.set_facecolor((0.9, 0.9, 0.9))
```
